## About Dataset


### Problem Statement

Customer Personality Analysis is a detailed analysis of a company’s ideal customers. It helps a business to better understand its customers and makes it easier for them to modify products according to the specific needs, behaviors and concerns of different types of customers.

Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company’s database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment.

### Column Information

#### People

ID: Customer's unique identifier

Year_Birth: Customer's birth year

Education: Customer's education level

Marital_Status: Customer's marital status

Income: Customer's yearly household income

Kidhome: Number of children in customer's household

Teenhome: Number of teenagers in customer's household

Dt_Customer: Date of customer's enrollment with the company

Recency: Number of days since customer's last purchase

Complain: 1 if the customer complained in the last 2 years, 0 otherwise



### Products

MntWines: Amount spent on wine in last 2 years

MntFruits: Amount spent on fruits in last 2 years

MntMeatProducts: Amount spent on meat in last 2 years

MntFishProducts: Amount spent on fish in last 2 years

MntSweetProducts: Amount spent on sweets in last 2 years

MntGoldProds: Amount spent on gold in last 2 years

### Promotion

NumDealsPurchases: Number of purchases made with a discount

AcceptedCmp1: 1 if customer accepted the offer in the 1st campaign, 0 otherwise

AcceptedCmp2: 1 if customer accepted the offer in the 2nd campaign, 0 otherwise

AcceptedCmp3: 1 if customer accepted the offer in the 3rd campaign, 0 otherwise

AcceptedCmp4: 1 if customer accepted the offer in the 4th campaign, 0 otherwise

AcceptedCmp5: 1 if customer accepted the offer in the 5th campaign, 0 otherwise

Response: 1 if customer accepted the offer in the last campaign, 0 otherwise

### Place

NumWebPurchases: Number of purchases made through the company’s website

NumCatalogPurchases: Number of purchases made using a catalogue

NumStorePurchases: Number of purchases made directly in stores

NumWebVisitsMonth: Number of visits to company’s website in the last month

```python
# Function to count outliers using IQR method
def count_outliers(column):
    if pd.api.types.is_numeric_dtype(column):
        Q1 = column.quantile(0.25)
        Q3 = column.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = column[(column < lower_bound) | (column > upper_bound)]
        return len(outliers)
    else:
        return 0

# Loop through each column and print the number of outliers
for column in df.columns:
    num_outliers = count_outliers(df[column])
    print(f"{column} outliers: {num_outliers}")
```

```python
df.isnull().sum()
```

```python
# Function to remove outliers using IQR method, excluding the label column
def remove_outliers(df):
    df_no_outliers = df.copy()
    for column in df.columns:
        if column != 'Response' and pd.api.types.is_numeric_dtype(df[column]):
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            df = df_no_outliers[(df_no_outliers[column] >= lower_bound) & (df_no_outliers[column] <= upper_bound)]
```

# MACHINE LEARNING

Splitting the dataset into test and train

```python
from sklearn.model_selection import train_test_split

# Define X (all columns except the target column)
X = df.drop('Response', axis=1)

# Define y (only the target column)
y = df['Response']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

### 1.Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, roc_curve, auc, precision_score,recall_score

# Initialize a Random Forest Classifier with 100 trees and a random state for reproducibility
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)

# Fit the Random Forest Classifier on the training data
rf_classifier.fit(X_train, y_train)

# Use the trained classifier to make predictions on the test data
RFC_y_pred = rf_classifier.predict(X_test)

# Calculate the performance on the test set
RFC_conf_matrix = confusion_matrix(y_test, RFC_y_pred)
RFC_accuracy = accuracy_score(y_test, RFC_y_pred)
RFC_f1 = f1_score(y_test, RFC_y_pred)
RFC_precision = precision_score(y_test, RFC_y_pred)
RFC_recall = recall_score(y_test, RFC_y_pred)

print("Confusion Matrix:\n", RFC_conf_matrix)
print("Accuracy:", RFC_accuracy)
print("F1 Score:", RFC_f1)
print("Precision:", RFC_precision)
print("Recall:", RFC_recall)
```

```python
# Use the trained Random Forest Classifier to predict class probabilities for the positive class (class 1) on the test data
RFC_y_prob = rf_classifier.predict_proba(X_test)[:, 1]

# Compute the false positive rate (fpr) and true positive rate (tpr) using the ROC curve function
fpr, tpr, _ = roc_curve(y_test, RFC_y_prob)

# Calculate the Area Under the Curve (AUC) for the Receiver Operating Characteristic (ROC) curve to assess model performance
RFC_roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(5, 3))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(RFC_roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

Why Random Forest Classifier?

Mixed Data Types: Random Forests can handle a mix of numerical and categorical features well. In the dataset, we have features like 'Education', 'Marital_Status', and 'Dt_Customer', which are categorical, and RandomForestClassifier can handle them without requiring extensive preprocessing.

Handling Imbalanced Data: Random Forests can handle imbalanced datasets well. If the dataset has a class imbalance (e.g., significantly more instances of one class than another), Random Forests can still perform effectively.

Feature Importance: Random Forests provide a feature importance score, which can help in understanding which features contribute the most to the model's predictions. This is valuable for interpreting the significance of different variables in the dataset. We find the scores:

### 4. Naive Bayes Classifier (NB)

The provided code outlines the implementation and evaluation of a Naive Bayes Classifier, a popular machine learning algorithm, using Python's scikit-learn library.

```python
from sklearn.naive_bayes import GaussianNB

# Initialize Naive Bayes Classifier
clf = GaussianNB()

# Training the classifier
clf.fit(X_train, y_train)

# Making predictions
NBC_y_pred = clf.predict(X_test)
NBC_y_prob = clf.predict_proba(X_test)[:, 1]

# Calculate and display evaluation metrics
NBC_conf_matrix = confusion_matrix(y_test, NBC_y_pred)
NBC_accuracy = accuracy_score(y_test, NBC_y_pred)
NBC_f1 = f1_score(y_test, NBC_y_pred)
NBC_precision = precision_score(y_test, NBC_y_pred)
NBC_recall = recall_score(y_test, NBC_y_pred)

# Evaluation metrics
print("Confusion Matrix:\n", NBC_conf_matrix)
print("Precision:",NBC_precision)
print("Recall:",NBC_recall)
print("Accuracy:", NBC_accuracy)
print("F1 Score:", NBC_f1)
```

Initialization of the Classifier: clf = GaussianNB() initializes the Gaussian Naive Bayes classifier. Gaussian Naive Bayes is suited for continuous data and assumes that the features follow a normal distribution.

```python
# Compute ROC curve and ROC area
fpr, tpr, _ = roc_curve(y_test, NBC_y_prob)
NBC_roc_auc = auc(fpr, tpr)

# Plotting ROC Curve
plt.figure(figsize=(5, 3))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(NBC_roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) for Naive Bayes')
plt.legend(loc="lower right")
plt.show()
```

#### When to use The Naive Bayes Classifier?

The Naive Bayes Classifier is particularly well-suited for scenarios where the assumption of feature independence holds true and computational efficiency is paramount. It shines in its simplicity and speed, making it ideal for large datasets and real-time prediction tasks. This method is a strong choice for text classification problems such as spam detection or sentiment analysis, where the features can be considered independent given the class label. Additionally, it performs well in multi-class prediction tasks and is often preferred when a probabilistic framework is required for the task at hand, owing to its foundation in Bayes' Theorem. It's a good baseline algorithm due to its straightforward implementation and can serve as a starting point for establishing the performance metrics before considering more complex models.

However, it might require more tuning and computational resources compared to simpler algorithms like logistic regression or decision trees.
